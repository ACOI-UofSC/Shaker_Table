```python
import glob
import pandas as pd
import numpy as np
```

```python
# path = "V:\\R01 - W4K\\2_Shaker project\\Devices Evaluated\\Reliability\\Processed Data\\Round 2\\Actigraph\\Trials\\"
# path = "V:\\R01 - W4K\\2_Shaker project\\Devices Evaluated\\Reliability\\Processed Data\\Round 2\\Apple\\Trials\\"
# path = "V:\\R01 - W4K\\2_Shaker project\\Devices Evaluated\\Reliability\\Processed Data\\Round 2\\Garmin\\Trials\\"
path = "V:\\R01 - W4K\\2_Shaker project\\Devices Evaluated\\BlackAccelerometerTest\\Processed Data\\Trials\\"
trials = glob.glob(path + "Trial [0-9][0-9].csv")
agg_trials = glob.glob(path + "Trial Aggregated [0-9][0-9].csv")
rms = lambda x : np.sqrt(np.mean(x ** 2))
```

```python

for trial_path, agg_path in zip(trials, agg_trials):
    # Read in aligned trial
    data = pd.read_csv(trial_path, parse_dates=["Time"], infer_datetime_format=True)
    # Calculate MEAN x,y,z, magnitude, and enmo
    data_agg = data.groupby(['Time', 'ID']).agg({"X": [np.mean], "Y": [np.mean], "Z": [np.mean], "Magnitude": [np.mean, rms], 
                                                 "ENMO": [np.mean, rms]}).reset_index()
    # Name columns
    data_agg.columns= ['Time', 'ID', 'MEAN X', 'MEAN Y', 'MEAN Z', 'MEAN Magnitude', 'RMS Magnitude', 'MEAN ENMO', 'RMS ENMO']
    
    # Read in aggregated trial
    old_agg = pd.read_csv(agg_path, parse_dates=['Time'], infer_datetime_format=True).iloc[:,:9]

    # Merge old aggregated data with newly calculated values
    new_agg = old_agg.merge(data_agg, how='left', on=['Time', 'ID'])

    
    # Write to file
    new_agg.to_csv(agg_path, index=False)
    
    
```
