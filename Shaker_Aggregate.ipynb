```python
import glob
import pandas as pd
```

```python
path = "V:\\R01 - W4K\\2_Shaker project\\Devices Evaluated\\Trial Data\\Processed data\\Garmin\\"
files = glob.glob(path + "Trial*//*rms.csv")
files
```

```python
data = None
for file in files:
    if data is None:
        data = pd.read_csv(file)
    else :
        temp = pd.read_csv(file)
        data = pd.concat([data, temp])    
data
```

```python
data.to_csv(path + "Garmin_RMS.csv", index=False)
```
